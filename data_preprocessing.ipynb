```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/dataset_PQ.csv")
data.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(data['label'].unique())
```

```python
le.transform(data['label'].unique())
```

```python
classes = dict(zip(le.classes_, le.transform(data['label'].unique())))
classes
```

```python
data['label'] = data['label'].map(classes)
```

```python
data['label'].unique()
```

```python
X_data = data[list(data.columns)[:-1]]
X_data.head()
```

```python
Y_data = data[['label']]
Y_data
```

```python
X_data.info()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_X_data = scaler.fit_transform(X_data)
```

```python
scaled_X_data_df = pd.DataFrame(scaled_X_data, columns=X_data.columns)
scaled_X_data_df.head()
```

```python
scaled_X_data_df.shape, Y_data.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_X_data_df, Y_data, random_state=42, test_size=0.1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train = X_train.sort_index()
X_test = X_test.sort_index()
y_train = y_train.sort_index()
y_test = y_test.sort_index()
```

```python
import pickle

exporte_data = {
    'dataset' : data,
    'scaled_X_data_df' : scaled_X_data_df,
    'y_data' : Y_data,
    'X_train' : X_train,
    'X_test' : X_test,
    'y_train' : y_train,
    'y_test' : y_test,
    'classes' : classes,
    'x_train_idx' : X_train.index,
    'y_train_idx' : y_train.index,
    'x_test_idx' : X_test.index,
    'y_test_idx' : y_test.index,
    'metadata' : {
        'test_size' : 0.1,
        'random_state' : 42
    }
}

with open('pq_data_preprocessing.pkl', 'wb') as f:
    pickle.dump(exporte_data, f)
```

```python
with open('pq_data_preprocessing.pkl', 'rb') as f:
  pq_data = pickle.load(f)
```
