```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv("/content/Autoencoder_results_64_train.csv")
train_data
```

```python
test_data = pd.read_csv("/content/Autoencoder_results_64_test.csv")
test_data
```

```python
merged_data = pd.concat([train_data, test_data], ignore_index=True)
merged_data
```

```python
merged_data.Label.unique()
```

```python
X = np.array(merged_data.drop('Label', axis=1))
X
```

```python
y = np.array(merged_data[['Label']])
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from RBFNN import RBFNN
import tensorflow as tf
```

```python
hidden_neurons = [11, 16, 21, 26, 32, 37, 42, 47, 53]
hidden_neurons
```

```python
import wandb
```

```python
!wandb login
```

```python
overall_metrics = {}
final_metrics = {}

epochs = 200

for n in hidden_neurons:
  for p in range(2, n+1, 2):
    bs = 32
    run = wandb.init(
        project = "RBFNN PQ Re_Run",
        name = f"N_{n}_P_{p}_BS_32",
        config = {
            "epochs" : epochs,
            "Neurons" : n,
            "Width Factor" : p,
            "Batch Size" : bs,
        }
    )

    rbf = RBFNN(X=X_train,
                y=y_train,
                X_test=X_test,
                y_test=y_test,
                n_hidden_neurons=n,
                width_factor=p,
                activation='softmax')
    
    rbf.fit(epochs=epochs, metrics=['accuracy'], loss_fn=tf.keras.losses.SparseCategoricalCrossentropy)

    overall_metrics[f'N_{n}_P_{p}_BS_{bs}'] = rbf.history.history

    final_metrics[f'N_{n}_P_{p}_BS_{bs}'] = {
        'accuracy' : overall_metrics[f'N_{n}_P_{p}_BS_{bs}']['accuracy'][-1],
        'val_accuracy' : overall_metrics[f'N_{n}_P_{p}_BS_{bs}']['val_accuracy'][-1],
        'loss' : overall_metrics[f'N_{n}_P_{p}_BS_{bs}']['loss'][-1],
        'val_loss' : overall_metrics[f'N_{n}_P_{p}_BS_{bs}']['val_loss'][-1]
    }

    wandb.config.update({
        'history' : rbf.history.history,
        'accuracy' : final_metrics[f'N_{n}_P_{p}_BS_{bs}']['accuracy'],
        'val_accuracy' : final_metrics[f'N_{n}_P_{p}_BS_{bs}']['val_accuracy'],
        'loss' : final_metrics[f'N_{n}_P_{p}_BS_{bs}']['loss'],
        'val_loss' : final_metrics[f'N_{n}_P_{p}_BS_{bs}']['val_loss']
    })

    wandb.save('/content/RBFNN.py')
    wandb.save(f"/content/Metadata_N_{n}_P_{p}_bs_{bs}.jbl")
    wandb.save(f"/content/RBF_ANN_model_bs_{bs}_N_{n}_P_{p}.h5")
```
