```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import sawtooth
from pywt import wavedec
from scipy.stats import skew, kurtosis, entropy

t = np.arange(0, 0.4, 0.0001)
alpha = 0.7
y = (1 - alpha * ((t >= 0.1) & (t <= 0.18))) * np.sin(314 * t)

plt.figure(2)
plt.plot(t, y)
plt.title('Sag disturbance')

coeffs = wavedec(y, 'db2', level=8)
A8 = coeffs[0]
d1, d2, d3, d4, d5, d6, d7, d8 = coeffs[1:]

mean_sp = [np.mean(d1), np.mean(d2), np.mean(d3), np.mean(d4), np.mean(d5), np.mean(d6), np.mean(d7), np.mean(d8), np.mean(A8)]
std_sp = [np.std(d1), np.std(d2), np.std(d3), np.std(d4), np.std(d5), np.std(d6), np.std(d7), np.std(d8), np.std(A8)]
rms_sp = [np.sqrt(np.mean(d1**2))*len(d1), np.sqrt(np.mean(d2**2))*len(d2), np.sqrt(np.mean(d3**2))*len(d3), np.sqrt(np.mean(d4**2))*len(d4), np.sqrt(np.mean(d5**2))*len(d5), np.sqrt(np.mean(d6**2))*len(d6), np.sqrt(np.mean(d7**2))*len(d7), np.sqrt(np.mean(d8**2))*len(d8), np.sqrt(np.mean(A8**2))*len(A8)]
range_sp = [np.max(d1) - np.min(d1), np.max(d2) - np.min(d2), np.max(d3) - np.min(d3), np.max(d4) - np.min(d4), np.max(d5) - np.min(d5), np.max(d6) - np.min(d6), np.max(d7) - np.min(d7), np.max(d8) - np.min(d8), np.max(A8) - np.min(A8)]
skew_sp = [skew(d1), skew(d2), skew(d3), skew(d4), skew(d5), skew(d6), skew(d7), skew(d8), skew(A8)]
kurtosis_sp = [kurtosis(d1), kurtosis(d2), kurtosis(d3), kurtosis(d4), kurtosis(d5), kurtosis(d6), kurtosis(d7), kurtosis(d8), kurtosis(A8)]
energy_sp = [(rms_sp[0]**2)*len(d1), (rms_sp[1]**2)*len(d2), (rms_sp[2]**2)*len(d3), (rms_sp[3]**2)*len(d4), (rms_sp[4]**2)*len(d5), (rms_sp[5]**2)*len(d6), (rms_sp[6]**2)*len(d7), (rms_sp[7]**2)*len(d8), (rms_sp[8]**2)*len(A8)]
entropy_sp = [entropy(d1), entropy(d2), entropy(d3), entropy(d4), entropy(d5), entropy(d6), entropy(d7), entropy(d8), entropy(A8)]
```
